```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Iris.csv")
```

```python
df.head(3)
```

make pair plot to see data is Linearly Seperable or Not

```python
sns.pairplot(data=df)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

Use Elbow Method to find How many clusers are formed in our Dataset

```python
wcss=[]
for i in range(2,21):
  kmean=KMeans(n_clusters=i,init='k-means++')
  kmean.fit(df)
  wcss.append(kmean.inertia_)
```

Make Graph Of Elbow Method

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xlabel("No Of Clusters")
plt.xticks([i for i in range(2,21)])
plt.ylabel("Wcss")
plt.grid(axis="x")
plt.show()
```

train the KMean Clustring

```python
kmean=KMeans(n_clusters=3)
kmean.fit_predict(df)
```

make predict column in dataset

```python
df['Predict']=kmean.fit_predict(df)
```

```python
df.head(3)
```

make graph to see the clusters

```python
sns.pairplot(data=df,hue='Predict')
plt.show()
```
