```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=250,noise=0.05)
```

```python
df={"data1":X[:,0],

    "data2":X[:,1],

    }
```

```python
dataset=pd.DataFrame(df)
```

```python
dataset.head(3)
```

make a scatter plot to check data is linearly seperable or not

```python
sns.scatterplot(x='data1',y='data2',data=dataset)
plt.show()
```

using DBSCAN clustring

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.2,min_samples=5)
db.fit_predict(dataset)
```

```python
dataset['predict']=db.fit_predict(dataset)
```

```python
dataset.head(3)
```

make a graph to see the clusters

```python
sns.scatterplot(x='data1',y='data2',data=dataset,hue='predict')
plt.show()
```
